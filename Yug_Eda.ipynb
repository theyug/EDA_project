```python
ax = sn.countplot( x='Gender' , data =df)
for bars in ax.containers:
  ax.bar_label(bars)
  #counterplot is a function to visaulize the datasets
```

```python
sales_gen =df.groupby(['Gender']  , as_index= False) ['Amount'].sum().sort_values(by='Amount' ,ascending =False)
sn.barplot( x='Gender', y='Amount', data = sales_gen)
#this data shows that females buy more
```

```python
ax = sn.countplot( data =df, x ='Age Group' , hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age =df.groupby(['Age Group']  , as_index= False) ['Amount'].sum().sort_values(by='Amount' ,ascending =False)
sn.set(rc={'figure.figsize': (15 ,5)})
sn.barplot( x='Age Group', y='Amount', data = sales_age)
```

```python
sales_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by ='Orders' , ascending = False)
sn.set(rc={'figure.figsize': (20 ,5)})
sn.barplot( x='State', y='Orders', data = sales_state)
#this plot show us that UP state is  highest in buying
```

```python
sales_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by ='Amount' , ascending = False)
sn.set(rc={'figure.figsize': (25 ,5)})
sn.barplot( x='State', y='Amount', data = sales_state)
#this plot show us that UP is the highest in buying
```

```python
ax = sn.countplot( x='Marital_Status' , data =df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status' , 'Gender'] , as_index = False)['Amount'].sum().sort_values(by ='Amount' , ascending = False)
sn.set(rc={'figure.figsize': (25 ,5)})
sn.barplot( x='Marital_Status', y='Amount',hue = 'Gender' , data = sales_state)
```

```python
ax = sn.countplot( data =df , x ='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
  #this plot shows that IT job people are high in count most
```

```python
sales_state = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by ='Amount' , ascending = False)
sn.set(rc={'figure.figsize': (25 ,5)})
sn.barplot( x='Occupation', y='Amount' , data = sales_state)
#this plot show us that  IT sector people are the highest in buying
```
